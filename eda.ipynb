# Make first Predictions

```python
X = data.drop('LUNG_CANCER',1)
y = data['LUNG_CANCER']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(X,y,test_size=0.2,random_state=26)
```

### lets do a regular linear regression

```python
from sklearn.linear_model import LogisticRegression,Lasso
```

```python
models = [LogisticRegression(),Lasso(),]
```

```python
best = 0
best_model = None
for m in models:
    m.fit(xtrain,ytrain)
    score = m.score(xtest,ytest)
    print(m,score)
    if score > best:
        best = score
        best_model = m
        
print(f"{best_model} has score pf {best}")
```
