```python
one2_col = ['SMOKING', 'YELLOW_FINGERS', 'ANXIETY',
       'PEER_PRESSURE', 'CHRONIC DISEASE', 'FATIGUE', 'ALLERGY', 'WHEEZING',
       'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH',
       'SWALLOWING DIFFICULTY', 'CHEST PAIN']
```

```python
for i in one2_col:
    data[i] = np.where(data[i] == 1,0,1)
```

```python
data.drop("GENDER",axis=1,inplace=True)
```

```python
data['is_male'] = np.where(data['GENDER'] == 'M',1,0)
```

```python
data['LUNG_CANCER'] = np.where(data['LUNG_CANCER'] == 'YES',1,0)
```

```python
data = data.drop(data[data["AGE"] < 30].index,0)
```
